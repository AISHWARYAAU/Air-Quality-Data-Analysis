## Data visualitation

```python
import matplotlib.pyplot as plt
```

```python
data.columns
```

```python
#train the data
df = pd.DataFrame(data)
```

```python
df = pd.DataFrame(data.head(20))
df['City'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("City Distribution")
plt.show()
```

```python
df = pd.DataFrame(data.head(20))
df['Rank'].plot(kind='hist', color='Blue')
plt.xlabel('Rank')
plt.ylabel('Count')
plt.title("Rank Distribution")
plt.show()
```

```python
df = pd.DataFrame(data.head(20))
df['Rank'].plot(kind='line', color='Blue', marker='o', linestyle='-')
plt.xlabel('Rank')
plt.ylabel('Count')
plt.title("Rank Distribution")
plt.show()
```

```python
df = pd.DataFrame(data.head(30))
plt.scatter(df.index, df['Rank'], s=100, alpha=0.7, edgecolors='black')
plt.xlabel('Index')
plt.ylabel('Rank')
plt.show()
```

```python
df = pd.DataFrame(data.head(30))
plt.scatter(df.index, df['City'], s=100, alpha=0.7, edgecolors='black')
plt.xlabel('Index')
plt.ylabel('City')
plt.show()
```

```python
df = pd.DataFrame(data.head(30))
plt.scatter(df.index, df['2023'], s=100, alpha=0.7, edgecolors='black')
plt.xlabel('Index')
plt.ylabel('2023 Air Quality Index')
plt.show()
```

```python
df = pd.DataFrame(data.head(20))
df['Country'].value_counts().plot(kind='bar', color='Green')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title("Country Distribution")
plt.show()
```

```python
df = pd.DataFrame(data.head(30))
df['Country'].value_counts().plot(kind='bar', stacked=True)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title("Country Distribution")
plt.show()
```

```python
#Generating a Box Plot for the 'Rank' Column
df = pd.DataFrame(data.head(30))
df['Rank'].plot(kind='box')
plt.xlabel('Rank')
plt.ylabel('Count')
plt.title("Rank Distribution")
plt.show()
```
